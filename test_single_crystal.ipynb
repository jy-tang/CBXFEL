```python
npad1 = (256 - ncar)//2
npadx = [int(npad1), int(npad1) + 1]
dfl_padx = pad_dfl_x(dfl, npadx)
fld_fft = np.fft.fftshift(fft2(dfl_padx, axes = (0,1)), axes=(0,1))
```

```python
nslice_padded = dfl.shape[0]
h_Plank = 4.135667696e-15      # Plank constant [eV-sec]
c_speed  = 299792458           # speed of light[m/sec]
dgrid = 540e-6
low = 9831.5
high = 9832.2

dt = xlamds*zsep /c_speed
nx = ny = ncar
nx_padded = ncar + int(npadx[0]) + int(npadx[1])
# get photon energy coordinate
hw0_eV = h_Plank * c_speed / xlamds
Dhw_eV = h_Plank / dt 
eph = hw0_eV + Dhw_eV / 2. * np.linspace(-1.,1., nslice_padded)
lambd = h_Plank*c_speed/eph

# get kx,ky coordinates
dx = 2. * dgrid / ncar
Dkx = 2. * np.pi / dx
kx = Dkx/ 2. * np.linspace(-1.,1.,nx_padded)

eph_cut = eph[(eph>low) &(eph < high)]


eph_mesh, kx_mesh = np.meshgrid(eph_cut, kx, indexing = 'ij')
spec = np.sum(np.abs(fld_fft)**2, axis = 2)
k0 = np.pi*2/xlamds  
spec = spec[(eph>low) &(eph < high),:]

fig = plt.figure(figsize = (10,10))
ax = fig.gca(projection='3d')
X = eph_mesh
Y = kx_mesh/k0
Z = spec
ax.plot_surface(X, Y, Z, rstride=8, cstride=8, alpha=0.3)
cset = ax.contourf(X, Y, Z, zdir='z', offset=0, cmap=cm.coolwarm)
cset = ax.contourf(X, Y, Z, zdir='x', offset=9831.5, cmap=cm.coolwarm)
cset = ax.contourf(X, Y, Z, zdir='y', offset=1.0e-5, cmap=cm.coolwarm)

ax.set_xlabel('photon energy (eV)', fontsize = 12)
ax.set_ylabel(r'$\theta_x$', fontsize = 12)
ax.set_zlabel('spectral intensity', fontsize = 12)


plt.show()

k0 = np.pi*2/xlamds
plt.plot(kx/k0, spec[spec.shape[0]//2, :], '-o')
plt.xlabel(r'$\Theta_x$')
plt.ylabel('Spectral Intensity')
plt.show()

plt.plot(eph_cut, spec[:, 128], '-o')
plt.xlabel('Photon energy (eV)')
plt.show()
```
