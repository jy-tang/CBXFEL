```python
theta_slice=np.pi/4
pi = np.pi
eph = np.linspace(9831.65,9832.15,5001);
theta_0 = 45.0*pi/180.; dtheta_0 = 50e-6; ntheta_0 = 101;
#theta_0 *= 0.
theta = np.linspace(theta_0-dtheta_0,theta_0+dtheta_0,ntheta_0);

R0H = Bragg_mirror_reflection(eph, theta)
R00 = Bragg_mirror_transmission(eph, theta)

Eph,Thetaurad = np.meshgrid(eph,1e6*(theta-theta_slice));
cut = Thetaurad == 0;

plt.plot(Eph[cut],np.abs(R0H[cut])**2, color = 'C0')

ax = plt.gca()
ax2 = ax.twinx()

sig = 0.02
m = 9831.89
p = np.exp(-(eph - m)**2/2/sig**2)
p /= np.max(p)

ax2.plot(eph, p, color = 'C2', label = 'chirper off' )

sig = 0.02
m = 9831.79
p = np.exp(-(eph - m)**2/2/sig**2)
p /= np.max(p)

ax2.plot(eph, p, color = 'C1' , label = 'chirper on')

ax2.legend(fontsize  =14)

ax.set_xlabel('Photon energy', fontsize = 14)
ax.set_ylabel('Bragg diffration intensity', color = 'C0', fontsize =14)
ax.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False, # labels along the bottom edge are off
    )
ax.tick_params(axis = 'y', colors = 'C0', labelsize = 14)
ax2.set_ylabel('Spectral Intensity', fontsize = 14, color ='C1')
ax2.tick_params(axis = 'y',  labelsize = 14)
plt.savefig('figures/layout_spec.png')
```
